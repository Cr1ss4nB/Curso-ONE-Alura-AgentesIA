# Funcionamiento de la Indexación Negativa

Cuando se usa un índice negativo, Python comienza a contar desde el final. Así, el índice -1 corresponde al último elemento, -2 al penúltimo, y así sucesivamente. Este conteo inverso es útil para acceder a elementos sin necesidad de saber el tamaño exacto de la lista. Por ejemplo:

```python
nombres = ['María', 'Juan', 'Ana', 'Pedro', 'Juliana']

print(nombres[-1])  # Muestra 'Juliana'
print(nombres[-2])  # Muestra 'Pedro'


# Ejemplo de slicing con índice negativo
ultimos_tres = nombres[-3:]
print(ultimos_tres)  # Muestra ['Ana', 'Pedro', 'Juliana']
```

```python
def resumir_emails(cuerpos_emails):
  lista_resumenes = []
  for numero, email in enumerate(cuerpos_emails):
    respuesta = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {
                "role": "user",
                "content": f"""Te voy a enviar el cuerpo de un correo electrónico. Quiero que lo resumas en una sola línea indicando la intención
                de dicho correo. El correo es el siguiente: {email}"""
            }
        ])
    texto = respuesta.choices[0].message.content
    print(f"E-Mail N° {numero + 1}: {texto}")
    lista_resumenes.append(f"E-Mail N° {numero + 1}: {texto}")
  return lista_resumenes
```

1. Criar um arquivo .txt a partir de uma lista de perguntas vindas de uma lista em Python [].

3. Obter respostas de um LLM para cada uma.

```python
# Etapa 1
lista_de_preguntas = [
    "¿Cuál es la capital oficial de Australia?",
    "¿Qué instrumento se utiliza para medir la presión atmosférica?",
    "¿Cuántos huesos tiene el cuerpo humano de un adulto?",
    "¿En qué año comenzó la Primera Guerra Mundial?",
    "¿Cuál es el océano más grande y profundo de la Tierra?"
]
```

```python
with open("preguntas.txt", "w", encoding="utf-8") as archivo:
    for pregunta in lista_de_preguntas:
        archivo.write(pregunta + "\n")
```

```python
# Etapa 2
lista_desafio = []

with open("preguntas.txt", "r", encoding="utf-8") as archivo:
    for linea in archivo:
        lista_desafio.append(linea.strip())

print(lista_desafio)
```

```python
# Etapa 3
lista_de_diccionarios_de_respuestas = []

for pregunta in lista_desafio:
    respuesta = client.chat.completions.create(
        model="llama-3.3-70b-versatile",
        messages=[
            {
                "role": "user",
                "content": f"""Te voy a enviar una pregunta y tú respondes con una respuesta corta. Pregunta: {pregunta}"""
            }
        ]
    )

    texto = respuesta.choices[0].message.content
    lista_de_diccionarios_de_respuestas.append({"pregunta":pregunta, "respuesta":texto})
```

```python
lista_de_diccionarios_de_respuestas
```

```python
# Etapa 4
with open("respuestas1.csv", "w", encoding="utf-8") as archivo:
    archivo.write("pregunta,respuesta\n")
    for diccionario in lista_de_diccionarios_de_respuestas:
        archivo.write(f"{diccionario['pregunta']},{diccionario['respuesta']}\n")
```

```python
# Etapa 5

df = pd.DataFrame(lista_de_diccionarios_de_respuestas)
df.to_csv("respuestas1.csv", index=False, encoding="utf-8")
```
